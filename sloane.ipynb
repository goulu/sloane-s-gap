# Sloane's Gap

distribution of the occurences of integer numbers in [The On-Line Encyclopedia of Integer Sequences (OEIS)](https://oeis.org/).

```python
import pandas as pd
from Goulib.math2 import is_prime, is_power, divisors
from Goulib.itertools2 import ilen
df = pd.read_csv('acra.csv', delimiter=';', quotechar='"')
xx=[]; yy=[]; #properties of number
px=[]; py=[]; #primes
fx=[]; fy=[]; #powers
hx=[]; hy=[]; #powers
x=0
f=0
for index, row in  df.iterrows():
    if index==10000:break
    x=int(row[0]); y=int(row[1])
    if is_prime(x):
        px.append(x);
        py.append(y);
    else: 
        h=ilen(divisors(x))
        if h>f:
            f=h
            hx.append(x);
            hy.append(y);
        else:
            if is_power(x):
                fx.append(x);
                fy.append(y);
            else:
                xx.append(x);
                yy.append(y);
```

```python
%matplotlib inline
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
import matplotlib.pyplot as plt
import numpy as np



plt.semilogy(xx,yy,'.', color='blue', markersize=1);
plt.semilogy(hx,hy,'.', color='yellow', markersize=1);
plt.semilogy(fx, fy, '.', color='green', markersize=1);
plt.semilogy(px, py, '.', color='red', markersize=1);
```
